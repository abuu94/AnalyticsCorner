```python
# Import the seaborn library
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

# Load a sample dataset 

```python
df = sns.load_dataset("tips") # Loads the tips dataset
# View the head of my data
df.head() # Firs 5 rows of my data
```

```python
df.isnull().sum()
```

# Basic Plotting functions
## Scatter Plot

```python
sns.scatterplot(x = "total_bill", y = "tip", data= df)
plt.show()
```

## Line plot

```python
sns.lineplot(x = "size", y = "tip", data=df)
plt.show()
```

## Bar Plot

```python
sns.barplot(x = "day", y ="total_bill", data = df)
plt.show()
```

# Statistical Plots

## Box Plot

```python
sns.boxplot(x = "day", y = "total_bill", data=df)
plt.show()
```

## Violin Plot

```python
sns.violinplot(x = 'day', y = 'total_bill', data = df)
plt.show
```

## Histogram (Distribution plot)

```python
sns.histplot(df['total_bill'], bins = 20)
plt.show()
```

## Kernel Density Estimation (KDE)

```python
##
sns.kdeplot(df['total_bill'], shade = True)
plt.show()
```

# Categorical Plots
## Count plots


```python
df
```

```python
sns.countplot(x = 'sex', data = df)
plt.show()
```

## Swarm plots

```python
sns.swarmplot(x = 'day', y = "total_bill", data = df)
plt.show()
```

# Correlation and Heatmaps
## Correlation Matrix

```python
df
```

```python
df2 = df[["total_bill", 'tip', 'size']]
df2
```

```python
corr = df2.corr()
sns.heatmap(corr, annot = True, cmap = 'coolwarm')
plt.show()
```

## Pairplot

```python
sns.pairplot(df)
plt.show()
```

# Customization in Seaborn
## Change color pellets

```python
sns.set_palette("pastel")
sns.boxplot(x = "day", y = "total_bill", data = df)
plt.show()
```

## Adding title and labels

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x = "total_bill", y = 'tip', data = df)
plt.title("Total bill vs Tip")
plt.xlabel("total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
```

## Changing themes


```python
sns.set_style("whitegrid")
sns.boxplot(x = "day", y = "total_bill", data = df)
plt.show()
```

```python
# Regression plots
sns.lmplot(x = "total_bill", y = 'tip', data = df)
plt.show()
```
