```python
import matplotlib.pyplot as plt
```

# Basic Ploting
## Line plot


```python
x = [1,2,3,4,5]
y = [10, 20, 30, 40, 50]
```

```python
# A simple line plot
plt.plot(x, y)
plt.show()
```

```python
# Better the line plot
plt.plot(x, y, marker = "o", linestyle = '--', color = 'red')
plt.title("A simple line plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
#Scatter plot
plt.scatter(x, y, color = "green")
plt.title(" A simple Scatter plot")
plt.show()
```

```python
# Bar chat
categories = ["A", "B", "C", "D"]
values = [4, 7, 5, 2]
plt.bar(categories, values, color = 'g')
plt.title("Simple Bar Chart")
plt.show()
```

### Advanced Matplotlib features

#### SubPlots

```python
fig, axes = plt.subplots(1,2, figsize = (6, 3))
axes[0].plot(x, y, "r")
axes[0].set_title("Line Chart")
axes[1].bar(categories, values, color = 'b')
axes[1].set_title("bar Plot")
plt.show()
```
